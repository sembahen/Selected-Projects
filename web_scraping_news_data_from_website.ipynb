```python
#importing libraries
import numpy as np
import pandas as pd
from pprint import pprint
```

Using requests to fetch the HTML code from the website

```python
import requests
url = 'https://www.space.com/news'
response = requests.get(url)
#make sure we have a valid response
if(response.ok):
 data= response.text
print(data)
```

NB: The code above executes a very long text of information extracted from the news website in raw form. For purposes of illustration I trucated it. If interested in , look at this version (https://drive.google.com/file/d/13vtYpnSuzIrz2gf4sbpW7Xhxd9MvXK9o/view?usp=sharing)

Using beautiful soup to import the library, create an object with the response text.

```python
# Beautiful soup
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
print(soup.prettify())
```

I identified that all the objects I needed were under 'div' and classname = 'content'. So I used <b>.find_all</b> to extract that information.

```python
summary = soup.find_all('div', class_='content')
summary
```

I created the algorithm below to obtain the specific text ie. time, headline, synopsis, datetime. 

```python
summary = soup.find_all('div', class_='content')
questions=[]
for question in summary:
  headline=question.select_one('h3').get_text()
  author=question.select_one('.by-author > span').get_text()
  synopsis=question.select_one('.synopsis').get_text()
  datetime = question.select_one('time').attrs['data-published-date']
  new_question={'headline':headline, 'author':author ,'synopsis':synopsis, 'datetime':datetime}
  questions.append(new_question)
pprint(questions)
```

The data is now parsed into dictionary type and is ready for further analysis/ interpretation.
